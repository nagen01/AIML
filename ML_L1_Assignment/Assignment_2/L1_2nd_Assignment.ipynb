```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ex_data = pd.read_csv('ex1data1.csv', sep=',', names=["City Name", "Food Truck Profit"])
```

```python
X = ex_data.iloc[:,0:1].values
y = ex_data.iloc[:,1:2].values
```

```python
plt.scatter(X,y,marker='o')
plt.xlabel("City Polulation")
plt.ylabel("Profit of Food Truck")
plt.title("City Population vs Food truck Profit")
plt.show()
```

```python
#Splitting data into training and test:
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#Visualizing Training data:
plt.scatter(X_train,y_train,Color = 'green')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.xlabel("City Polulation")
plt.ylabel("Profit of Food Truck")
plt.title("Training : City Population vs Food truck Profit")
plt.show()
```

```python
#Visualizing Testing data:
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.xlabel("City Polulation")
plt.ylabel("Profit of Food Truck")
plt.title("Testing: City Population vs Food truck Profit")
plt.show()
```
