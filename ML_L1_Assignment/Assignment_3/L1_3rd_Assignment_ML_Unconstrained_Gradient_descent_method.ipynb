```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading data:
mpg_data = pd.read_csv('auto-mpg.data',sep='\s+',
                       names = ['mpg','cylinders','displacement','horsepower','weight','accelaration','model_year','origin','car name'])
```

```python
X = mpg_data.iloc[:,5:6].values
y = mpg_data.iloc[:,0].values
```

```python
plt.scatter(X,y)
plt.show()
```

```python
#Target is to find best fit line y = mx + c
#we need to calculate values of m and c 

alpha = 0.0001 # Learning Rate
epochs = 10000 # Number of iterations to perform the gradient descent

def linreg_gradient(X,y,alpha,epochs):
    m = 0 # Slope
    c = 0 # y intercept

    n = float(len(X))  # number of elements in X

    for i in range(epochs):
        y_pred = m*X + c #Current predicted value of y
        D_m = (-2/n) * sum(X * (y - y_pred)) #Derivative with respect to m
        D_c = (-2/n) * sum(y - y_pred) #Derivative with respect to c
        m = m - alpha * D_m # m update
        c = c - alpha * D_c # c update
    
#print(m,c)
```

```python
alpha = 0.0001  
epochs = 10000

linreg_gradient(X,y,alpha,epochs)

y_pred = m*X + c

X_min = X[:,0].min()
X_max = X[:,0].max()
y_pred_min = y_pred[:,0].min()
y_pred_max = y_pred[:,0].max()

plt.scatter(X,y)
plt.plot()
plt.plot(np.array([X_min, X_max]), np.array([y_pred_min, y_pred_max]), color = 'red') # Regression Line
plt.show()
```

```python
# Changing learning rate and epochs
alpha = 0.001
epochs = 20000

linreg_gradient(X,y,alpha,epochs)

y_pred = m*X + c

X_min = X[:,0].min()
X_max = X[:,0].max()
y_pred_min = y_pred[:,0].min()
y_pred_max = y_pred[:,0].max()

plt.scatter(X,y)
plt.plot()
plt.plot(np.array([X_min, X_max]), np.array([y_pred_min, y_pred_max]), color = 'red') # Regression Line
plt.show()
```
